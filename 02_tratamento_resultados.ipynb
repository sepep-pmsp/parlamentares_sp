## Tratamento resultados

Nesse notebook é feito um tratamento na base de dados para filtrar apenas os deputados eleitos

```python
from utils.load_csv import load_csv
```

```python
resultados_2022 = load_csv('resultados_2022.csv')
```

```python
filtro = (resultados_2022['DS_CARGO'].isin(['Deputado Federal', 'Deputado Estadual']) & resultados_2022['DS_SIT_TOT_TURNO'].isin(['ELEITO POR QP', 'ELEITO POR MÉDIA']))

deputados_eleitos = resultados_2022.loc[filtro].copy()
```
