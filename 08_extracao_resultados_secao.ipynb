## Extração resultados seção

Nesse notebook é extraído o resultado das eleições por seção eleitoral

```python
import pandas as pd
import os
import zipfile
import requests
from io import BytesIO
from utils.save_csv import save_csv
```

```python
file_tse_2022 = 'resultados_2022_secao.csv'

url_tse_2022 = 'https://cdn.tse.jus.br/estatistica/sead/odsele/votacao_secao/votacao_secao_2022_SP.zip'
```

```python
if os.path.exists(file_tse_2022):

    resultados_2022 = pd.read_csv(file_tse_2022, 
                                  sep = ';', 
                                  encoding = 'latin1')

else:

    response = requests.get(url_tse_2022)

    with zipfile.ZipFile(BytesIO(response.content)) as z:

        csv_name = [f for f in z.namelist() if 'SP' in f and f.endswith('.csv')][0]

        with z.open(csv_name) as f:

            resultados_2022 = pd.read_csv(f, 
                                          sep = ';', 
                                          encoding = 'latin1')

    save_csv(resultados_2022, file_tse_2022)
```
