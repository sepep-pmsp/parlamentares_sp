## Votos distrito

Nesse notebook os votos dos parlamentares são agregados por distrito

```python
import pandas as pd
from utils.load_csv import load_csv
from utils.save_csv import save_csv

pd.set_option('display.max_columns', None)
```

```python
dados_parlamentares = load_csv('dados_parlamentares.csv')

resultados_2022_secao = load_csv('resultados_2022_secao.csv', sep=';')

locais_votacao_geoloc_distritos = load_csv('locais_votacao_geoloc_distritos.csv', sep=';')
```

```python
resultados_distrito = resultados_2022_secao.merge(
    locais_votacao_geoloc_distritos[['DS_LOCAL_VOTACAO_ENDERECO', 'distrito']],
    on='DS_LOCAL_VOTACAO_ENDERECO',
    how='left'
)
```

```python
resultados_distrito = resultados_distrito.merge(
    dados_parlamentares[['id_deputado', 'nome_urna', 'partido']],
    left_on='SQ_CANDIDATO',
    right_on='id_deputado',
    how='left'
)
```

```python
resultados_distrito['id_deputado'] = (
    resultados_distrito['id_deputado']
        .astype('Int64')
        .astype(str)
)
```

```python
votos_por_distrito = (
    resultados_distrito
        .groupby(['id_deputado', 'nome_urna', 'partido', 'distrito'])
        .agg(qt_votos=('QT_VOTOS', 'sum'))
        .reset_index()
)
```
