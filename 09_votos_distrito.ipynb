## Votos distrito

Nesse notebook os votos dos parlamentares são agregados por distrito

```python
import pandas as pd
from utils.load_csv import load_csv
from utils.save_csv import save_csv

pd.set_option('display.max_columns', None)
```

```python
dados_parlamentares = load_csv('dados_parlamentares.csv', sep=';', engine='python')

resultados_2022_secao = load_csv('resultados_2022_secao.csv', sep=';', engine='python')

locais_votacao_geoloc_distritos = load_csv('locais_votacao_geoloc_distritos.csv', sep=';', engine='python')
```

```python
resultados_distrito = resultados_2022_secao.merge(
    locais_votacao_geoloc_distritos[['DS_LOCAL_VOTACAO_ENDERECO', 'distrito']],
    on='DS_LOCAL_VOTACAO_ENDERECO',
    how='left'
)
```

```python
resultados_distrito = resultados_distrito.merge(
    dados_parlamentares[['id_deputado', 'nome_urna', 'partido']],
    left_on='SQ_CANDIDATO',
    right_on='id_deputado',
    how='left'
)
```

```python
votos_por_distrito = (
    resultados_distrito
    .groupby(['id_deputado', 'nome_urna', 'partido', 'distrito'])
    ['QT_VOTOS']
    .sum()
    .reset_index()
)
```

```python
save_csv(votos_por_distrito, 'votos_por_distrito.csv', sep=';', index=False)
```
